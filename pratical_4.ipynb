# 1. Complex multiplication

```python
def complex_mult(z, w):
    realpart = z[0]*w[0] - z[1]*w[1]
    imagpart = z[1]*w[0] + z[0]*w[1]

    return (realpart, imagpart)

def fc(c):
    return f'{c[0]} + {c[1]}i'

# testing
z = (1, -1)
w = (3, 4)
print(f'product of {fc(z)} and {fc(w)} is {fc(complex_mult(z,w))}')
z = (2, -1)
w = (-3, 2)
print(f'product of {fc(z)} and {fc(w)} is {fc(complex_mult(z,w))}')
z = (3, 2)
w = (2, -5)
print(f'product of {fc(z)} and {fc(w)} is {fc(complex_mult(z,w))}')
```

# 2. Prime number check 

```python
def is_prime_for(n):
    if n <= 0:
        print(f'{n} is not a natural number')
        return False
    elif n <= 2:
        print(f'{n} is a prime number.')
        return True
    else:
        for i in range(n-2):
            d = i+2
            if n % d == 0:
                print(f'{n} is not a prime number. It is divisible by {d}')
                return False
        print(f'{n} is a prime number.')
        return True

is_prime_for(49)
is_prime_for(41)
is_prime_for(-333)
is_prime_for(1)
is_prime_for(2)
```

```python
def is_prime_while(n):
    if n <= 0:
        print(f'{n} is not a natural number')
        return False
    elif n <= 2:
        print(f'{n} is a prime number.')
        return True
    else:
        d = 2
        while d < (n-1):
            if n % d == 0:
                print(f'{n} is not a prime number. It is divisible by {d}')
                return False
            d += 1
        print(f'{n} is a prime number.')
        return True

is_prime_while(49)
is_prime_while(41)
is_prime_while(-333)
is_prime_while(1)
is_prime_while(2)
```

# 3. Derivative

```python
def float_list(start, end, step):
    li = []
    c = start
    while c < end:
        li.append(c)
        c += step
    return li

def func(x):
    return (x**4) - 2*x*x

def deriv(x):
    h = 0.01

    return (func(x+h) - func(x))/h

def deriv2(x):
    h = 0.01

    return (deriv(x+h) - deriv(x))/h

import matplotlib.pyplot as plt
x = float_list(-2, 2, 0.01)
y = [func(i) for i in x]
plt.plot(x, y, label='y')

y = [deriv(i) for i in x]
plt.plot(x, y, label='y\'')

y = [deriv2(i) for i in x]
plt.plot(x, y, label='y"')

plt.legend()
plt.show()
```

# 4. Sum of digits

```python
def sum_digits(n):
    if int(n) != n:
        print(f'{n} is not an integer!')
        return -1
    sum = 0
    
    rem = n 
    while rem > 0:
        d = rem % 10
        #print(d)
        sum += d
        rem = int(rem/10)
        #print(rem)

    return sum

print(f'{sum_digits(1000000)}')
print(f'{sum_digits(1940332)}')
print(f'{sum_digits(-1234)}')
print(f'{sum_digits(-12.34)}')
```

# Challenge: Unitary and Hermitial Matrices

```python
def transpose(M):
    MT = [[0, 0],[0, 0]]
    for i in range(2):
        for j in range(2):
            MT[j][i] = M[i][j]
    return MT

def matrix_mult(M1, M2):
    mult =  [[0, 0],[0, 0]]
    for i in range(2):
        for j in range(2):
            for k in range(2):
                mult[i][j] += (M1[i][k] * M2[k][j])
    return mult

def is_identical_matrix(M1, M2):
    for i in range(2):
        for j in range(2):
            if M1[i][j] != M2[i][j]: return False
    return True

def is_identity_matrix(M):
    I = [[1,0],[0,1]]
    return is_identical_matrix(M, I)

def is_unitary(M):
    MT = transpose(M)
    print(MT)
    p = matrix_mult(M, MT)
    print(p)
    return is_identity_matrix(p)

def is_hermitian(M):
    MT = transpose(M)
    return is_identical_matrix(M, MT)

# test
sigma_x = [[0,1],[1,0]]
sigma_z = [[1, 0],[0, -1]]
print(f'Is sigma_x is Hermitian? {is_hermitian(sigma_x)}')
print(f'Is sigma_x is Unitary? {is_unitary(sigma_x)}')
print(f'Is sigma_z is Hermitian? {is_hermitian(sigma_z)}')
print(f'Is sigma_z is Unitary? {is_unitary(sigma_z)}')
print(f'Is  [[1, 2],[-2, -1]] is Hermitian? {is_hermitian( [[1, 2],[-2, -1]])}')
print(f'Is  [[1, 2],[-2, -1]] is Unitary? {is_unitary( [[1, 2],[-2, -1]])}')
```
