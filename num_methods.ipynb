```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def dydt(y, t):
    return  (y*np.sin(t))**2

def rk4(f, h, t0, tf, y0):
    y = y0
    t = t0
    ts = [0]
    ys = [y0]
    while t < tf:
        t += h
        k1 = h*f(t, y)
        k2 = h*f(t+h/2, y+k1/2)
        k3 = h*f(t+h/2, y+k2/2)
        k4 = h*f(t+h, y+k3)
        y += (1/6)*(k1+2*(k2+k3)+k4)
        ys.append(y)
        ts.append(t)
        
    return ts,ys

for steps in [256]:
    ts = np.linspace(0, 2*np.pi, steps)
    ts, ys = rk4(dydt, 2*np.pi/steps, 0, 2*np.pi, 1)
    derivs = [dydt(t, y) for t, y in zip(ts, ys)]
    plt.plot(ts, ys, marker='.', label='y')
    plt.plot(ts, derivs, label='dydt')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.legend()
    plt.show()

func, [0, 20], [n0, pe0], t_eval=t, method='LSODA', args=(N, E, Gd, Gu))
    sol = solve_ivp(dydt, [0,2*np.pi], [1])
    plt.plot(ts, ys, marker='.', label='y')
    plt.plot(ts, derivs, label='dydt')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.legend()
    plt.show()

```
