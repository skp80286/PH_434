# 1. Transverse field Hamiltonian

```python
import numpy as np


paulix = np.array([[0,1],[1,0]])
pauliz = np.array([[1,0],[0,-1]])
identity = np.array([[1,0],[0,1]])

N=8
J=1
h=1/2
term1=None
for j in range(N-1):
    term_j = None
    for i in range(N):
        term_i = None
        if (i == j or i == j+1):
            term_i = paulix
        else:
            term_i = identity
        
        if term_j is None:
            term_j = term_i
        else:
            term_j = np.kron(term_j, term_i)
    if term1 is None: term1 = term_j
    else: term1 += term_j

term2 = None
for j in range(N):
    term_j = None
    for i in range(N):
        term_i = None
        if (i == j):
            term_i = pauliz
        else:
            term_i = identity
        
        if term_j is None:
            term_j = term_i
        else:
            term_j = np.kron(term_j, term_i)

    if term2 is None: term2 = term_j
    else: term2 += term_j

##
J,h = 1,1/2
H = J*term1+h*term2
print(f'Hamiltonian: {H}')
eigenresult = np.linalg.eigh(H)
print(f'eigenresult: {eigenresult}')
```
